# **OUTLIERS**

**Detect the outlier**

```python
dat_set7.boxplot()
```

```python
dat_set7['E_Plug'].quantile(0.25) #Q1
```

```python
dat_set7['E_Plug'].quantile(0.75)#Q#
```

```python
IQR=((dat_set7['E_Plug'].quantile(0.75))-(dat_set7['E_Plug'].quantile(0.25)))
print(IQR)
```

**Find why there is a dot in the boxplot graph**

MILD OUTLIER

```python
#Lower Bound=Q1-1.5*IQR
lb=(dat_set7['E_Plug'].quantile(0.25))-(1.5*IQR)
##Upper Bound =Q3+1.5*IQR
ub=(dat_set7['E_Plug'].quantile(0.75))+(1.5*IQR)
print(lb, ub)
```

EXTREME OUTLIER

```python
lb_e=(dat_set7['E_Plug'].quantile(0.25))-(1.5*IQR)
ub_e=(dat_set7['E_Plug'].quantile(0.75))+(3*IQR)
print(lb_e, ub_e)
```
